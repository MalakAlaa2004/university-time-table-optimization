```python
courses = courses[['course_id', 'course']].drop_duplicates()
len(courses)
```

```python
import random

# Number of professors needed
num_professors = 46

# Sample first and last names
first_names = [
    "Ahmed", "Fatima", "John", "Sarah", "Ali", "Emma", "Omar", "Lina", "James", "Maya",
    "Noah", "Sophia", "William", "Isabella", "Elijah", "Olivia", "Abdullah", "Layla",
]

last_names = [
    "Smith", "Johnson", "Williams", "Brown", "Jones", "Garcia", "Miller", "Davis", "Wilson", "Anderson",
    "Thomas", "Taylor", "Moore", "Jackson", "Martin", "Lee", "Perez", "Thompson", "White", "Harris",
]

# Generate random doctor names
professors = [f"{random.choice(first_names)} {random.choice(last_names)}" for _ in range(num_professors)]

# Make sure names are unique
professors = list(set(professors))

# If duplicates reduced the number, generate again
while len(professors) < num_professors:
    professors.append(f"{random.choice(first_names)} {random.choice(last_names)}")

professors = professors[:num_professors]

print(professors)
```

```python
import pandas as pd

# Example course list

# Repeat the list of professors enough times
professors_repeated = professors * (466 // len(professors) + 1)

# Shuffle to randomize assignments
random.shuffle(professors_repeated)

# Assign
course_assignments = pd.DataFrame({
    'course_id': courses["course_id"],
    'name':courses["course"],
    'instructor': professors_repeated[:466]
})

print(course_assignments.head())

# Save to CSV
course_assignments.to_csv("course_data.csv", index=False)
```
